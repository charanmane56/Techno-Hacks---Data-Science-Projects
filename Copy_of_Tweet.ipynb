```python
df['overall_sentiment'] = np.where(df['airline_sentiment'] == 'positive',1,np.where(df['airline_sentiment'] == 'negative',-1,0))
```

```python
df['overall_sentiment_confidence'] = df['airline_sentiment_confidence'] - df['negativereason_confidence']
```

```python
print(df)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['overall_sentiment_confidence'],bins=20,alpha=0.7,color='blue')
plt.xlabel('overall sentiment confidence')
plt.ylabel('count')
plt.title('Distibution of Overall Sentiment Confidence')
plt.show()
```
